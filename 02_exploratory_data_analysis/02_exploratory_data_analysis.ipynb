## Предобработка данных

### `first_day_exposition` и `rooms`

Конвертируем столбец `first_day_exposition` в тип datetime64 методом `pd.to_datetime`

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Хоть это и не пропущенные данные, но исправляем нулевое количество комнат, меняем на минмальное возможное (т.е. одну)

```python
data.loc[data['rooms'] == 0, 'rooms'] = 1
```

### `ceiling_height`

Пропуски в `ceiling_height` заполним просто стандартным значением высоты потолков 2,75 метра, что близко к наблюдаемому медианному значению 2,65.

```python
data['ceiling_height'].describe()
```

```python
data['ceiling_height'].fillna(2.75, inplace=True)
data['ceiling_height'].describe()
```

### `floors_total`

Пропуски в `floors_total` заполним значением из колонки `floor`, так как очевидно, что в здании не может быть меньше этажей, чем этаж, на котором расположена квартира, также меняем тип значений на int64 (целочисленные)

```python
data['floors_total'].fillna(data['floor'], inplace=True)
data['floors_total'] = data['floors_total'].astype('int64')
```

### `living_area`

Пропуски в `living_area` заполним медианным значением в зависимости от количества комнат `rooms` с использованем лямбда функции.

```python
data['living_area'] = data.groupby('rooms')['living_area'].transform(lambda x: x.fillna(x.median()))
```

Но в это случае возникают ошибочные данные в виде значений `living_area` больших, чем `total_area`, что неверно по своей сути.

```python
len(data.query('total_area - living_area < 0'))
```

В связи с этим используем функционал pandas под названием `where`, для замены полученного значения на максимальное возможное (total_area)

```python
data['living_area'].where(
    (data['total_area'] > data['living_area']),
    other = data['total_area'],
    inplace = True
)
```

```python
len(data.query('total_area - living_area < 0'))
```

### `is_apartment`

Пропусков в `is_apartment` слишком большое количество (20924 пропуска), так что опираться на усреднённые значения не представляется возможным, но и отбросить их нельзя. Следовательно принимаем все пропущенные значения как `False` в связи с фактической меньшей долей распространением апартаментов среди всего жилого фонда. Плюс в реальной ситуации желательно было бы запросить уточнение у составителей данной базы.

```python
data['is_apartment'].fillna(False, inplace=True)
```

### `kitchen_area`

Пропуски в `kitchen_area` заполним медианным значением в зависимости от количества комнат `rooms` с использованем лямбда функции, по аналогии с заполнением пропусков в `living_area`.

```python
data['kitchen_area'] = data.groupby('rooms')['kitchen_area'].transform(lambda x: x.fillna(x.median()))
```

### `balcony` и `locality_name`

Пропущенные значения количества балконов `balcony` (исходя из большого количества подобных пропусков) предположим как равные нулю. (т.е. если балконов нет, то их количество скорее всего и не заполняли, что вызвало пропуски в данных), и преобразуем в целочисленные значения.

```python
data['balcony'] = data['balcony'].fillna(0).astype('int64')
```

Пропуски в `locality_name` восстановить весьма сложно, в связи с чем просто удалим данные строки.

```python
data.dropna(subset=['locality_name'], inplace=True)
```

В связи с большим количеством пропусков в графах `airports_nearest`, `cityCenters_nearest` и `parks_around3000`, `ponds_around3000`, и невозможностью их восполнить, оставляем их без изменения.

Пропуски в `days_exposition` оставим так же без изменений.

### Редкие и выбивающиеся значения

Мы можем начать фильтрацию на редкие и выбивающиеся значения основываясь на полученных гистограммах из предыдущего пункта.

Но для более точного определения максимальных и минмальных значений, по которым мы будем обрезать нашу таблицу, напишем простую функцию, которая будет выдавать в цифровом виде результат, который графически отображается в виде "ящика с усами", по той же самой зависимости от межквартильного размаха.

```python
def iqr_max(column):
    iqr = column.quantile(q=0.75) - column.quantile(q=0.25)
    maxi = column.quantile(q=0.75)+1.5*iqr
    if maxi <= column.max():
        return maxi
    else:
        return column.max()        
        
def iqr_min(column):
    iqr = column.quantile(q=0.75) - column.quantile(q=0.25)
    mini = column.quantile(q=0.25)-1.5*iqr
    if mini >= column.min():
        return mini
    else:
        return column.min()
```

Пример работы моих функций `iqr_max` и `iqr_min`:

```python
print('Максимальное значения столбца {} составляет {}'
      .format(data['last_price'].name, iqr_max(data['last_price'])))
print('Минимальное значения столбца {} составляет {}'
      .format(data['last_price'].name, iqr_min(data['last_price'])))
```

Теперь применим эту фукнцию в вызове метода `.loc`, так как выяснилось, что `query` не поддерживает вызов функций

```python
good_data = data.loc[(data['last_price'] >= iqr_min(data['last_price'])) & 
                     (data['last_price'] <= iqr_max(data['last_price'])) &
                     (data['total_area'] >= iqr_min(data['total_area'])) & 
                     (data['total_area'] <= iqr_max(data['total_area'])) &
                     (data['rooms'] >= iqr_min(data['rooms'])) & 
                     (data['rooms'] <= iqr_max(data['rooms'])) &
                     (data['ceiling_height'] >= iqr_min(data['ceiling_height'])) & 
                     (data['ceiling_height'] <= 4)]
```

В результате чего, получаем обновлёный DataFrame уменьшенного размера, суммарные потери (процент отброшенных строк) составит:

```python
print(round(((data.shape[0] - good_data.shape[0]) / data.shape[0])*100, 2), '%')
```
