#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Создание новых признаков

Создадим копию DataFrame для дальнейших изменений с ресемплирование в 1 час в соответствии с заданием:

```python
orders_df = data.resample('1H').sum().copy()
```

Добавим такие календарные признаки, как день недели, месяц и число:

```python
orders_df['dayofweek'] = orders_df.index.dayofweek
orders_df['hour'] = orders_df.index.hour
```

Добавим новый признак методом "скользящего окна". Для начала определим его размер.

Количество записей, соответсвующее одному дню составляет (на примере одного месяца):

```python
daily_records = int(orders_df['2018-05-01':'2018-05-31'].shape[0]/31)

print(daily_records, 'записи в день')
```

На основе этого добавим признак **rolling_mean**

```python
orders_df['rolling_mean'] = orders_df['num_orders'].shift().rolling(daily_records).mean()

orders_df[['num_orders', 'rolling_mean']].plot()
plt.legend(['Количество заказов', 'Скользящее среднее'])
plt.title('График заказов такси от времени')
plt.show()
```

Добавим **отстающие значения**, их количество также подберём на основе количества записей за день.

```python
for i in range(1, daily_records+1, 11):
        orders_df[f'lag_{i}'] = orders_df['num_orders'].shift(i)
```

И отбросим образовавшиеся пропуски:

```python
orders_df = orders_df.dropna()
```

## Обучение

Разделим выборку на тренировочную и тестовую в соотношении $\frac{90\%}{10\%}$:

```python
X_train, X_test = train_test_split(orders_df, shuffle=False, test_size=0.1)
```

```python
y_train = X_train.pop('num_orders')
y_test = X_test.pop('num_orders')
```

Результаты работы всех моделей будут собраны в новый DataFrame()

```python
results = pd.DataFrame(columns=['RMSE', 'fit_time', 'predict_time'])
```

Добавим в эту таблицу как baseline предсказание предыдущим значением:

```python
pred_previous = y_test.shift()
pred_previous.iloc[0] = y_train.iloc[-1]

results.loc['Previous'] = [mean_squared_error(y_test, pred_previous)**0.5, 0, 0]
```

Добавим функцию, заполняющую таблицу с результатами. Измерение времени будет выполняться с использованием `datetime.now()`:

```python
def record(model, result_df=results,
           X=X_train,
           y=y_train,
           X_tru=X_test,
           y_tru=y_test):
    model_name = model.__class__.__name__
    
    time_start = datetime.now()
    model.fit(X, y)
    time_mid = datetime.now()
    y_pred = model.predict(X_tru)
    rmse = (mean_squared_error(y_tru, y_pred))**0.5
    time_end = datetime.now()
    results.loc[model_name] = [rmse,
                               (time_mid - time_start).total_seconds(),
                               (time_end - time_mid).total_seconds()]
    print(f'Model {model_name}, RMSE = {rmse}')
    return model, y_pred
```

### DummyRegressor()

Первым делом создадим baseline модель `DummyRegressor()`. 

```python
dummy_model = DummyRegressor()

dummy_model, dummy_pred = record(model=dummy_model)
```

### LinearRegression()

Проверим модель линейной регрессии

```python
linear_model = LinearRegression()

linear_model, linear_pred = record(model=linear_model)
```

### CatBoostRegressor()

Воспользуемся одним из вариантов реализации градиентного спуска:

```python
cat_model = CatBoostRegressor(learning_rate=0.03,
                              depth=6,
                              loss_function='RMSE',
                              random_seed=13,
                              verbose=50)

cat_model, cat_pred = record(model=cat_model)
```

### lightgbm()

Второй вариант градиентного спуска:

```python
lgb_model = lgb.LGBMRegressor(
    boosting_type='gbdt',
    max_depth=-1,
    learning_rate=0.5,
    n_estimators=100,
    random_state=13,
    n_jobs=-1
)

lgb_model, lgb_pred = record(model=lgb_model)
```

### RandomForestRegressor()

Для модели случайного леса в начале необходимо определить наилучшие гиперпараметры с использование `GridSearchCV()`, в связи с характером данных (время-зависимые) в качестве параметра `cv=` используем `TimeSeriesSplit`:

```python
rfr_clf = RandomForestRegressor()
rfr_parameters = {'n_estimators': range(100, 251, 25),
                  'max_depth': range(10, 20, 2),
                  'random_state': [13],
                  'n_jobs': [-1]
                 }
```

Для использования **RMSE** создадим свой собственный измеритель

```python
def RMSE(y_true, y_pred):
    return (mean_squared_error(y_true, y_pred)**0.5)

scorer_rmse = make_scorer(RMSE, greater_is_better=False)
```

```python
%%time
tscv = TimeSeriesSplit(n_splits=8)

grid_rfr = GridSearchCV(rfr_clf,
                        rfr_parameters,
                        scoring=scorer_rmse,
                        cv=tscv.split(X_train))
grid_rfr.fit(X_train, y_train)
grid_rfr.best_params_
```

```python
rfr_model = RandomForestRegressor(**grid_rfr.best_params_)

rfr_model, rfr_pred = record(model=rfr_model)
```

## Тестирование

### Анализ результатов

Рассмотрим полученные результаты:

```python
results
```

По результатам данной таблицы:

- Все модели проходят проверку на адекватность, показывая лучший результать чем `DummyRegressor()` и предыдущем значением;
- Наилучший результат в данном случае показал `CatBoostRegressor()`, со своим значением **RMSE = 41.93**;
- Второе место по результату **RMSE = 42.47** имеет `LGBMRegressor()`, но с учётом времени работы (в 11 раз быстрее чем CatBoost), является наилучшим выбором для внедрения;
- Самый быстрый результат, проходящий проверку на адекватность и удовлетворяющий условиям - `LinearRegression()` с результатом **RMSE = 46.57**.

Сравним графически полученные результаты:

```python
plt.figure(figsize=(17, 10))

y_test.plot(color='black', style='--', alpha=0.2)
pd.Series(dummy_pred, index=y_test.index).plot(ax=plt.gca(), alpha=0.5, style=':')
pd.Series(linear_pred, index=y_test.index).plot(ax=plt.gca(), alpha=0.5, style=':')
pd.Series(cat_pred, index=y_test.index).plot(ax=plt.gca(), alpha=0.5, style=':')
pd.Series(lgb_pred, index=y_test.index).plot(ax=plt.gca(), alpha=1, style='-')
pd.Series(rfr_pred, index=y_test.index).plot(ax=plt.gca(), alpha=0.5, style=':')

plt.legend(['Истинное значение',
            'DummyRegressor',
            'LinearRegression',
            'CatBoostRegressor',
            'LGBMRegressor',
            'RandomForestRegressor'           
           ])
plt.title('Количество заказов в час')
plt.show()
```

**Промежуточный вывод:**

Характер графиков результатов подтверждает, что общие тенденции на рост и падение определили верно все модели, кроме `DummyRegressor()`.

Также это демонстрирует проверку на адекватность, не только в виде разницы между $RMSE_{dummy}=84.6$ и $RMSE_{lgb}=42.5$. Ожидаемо, результат DummyRegressor совершенно не показывает колебания спроса, и является не просто "моделью хуже в два раза", а совершенно непригодной моделью (в чем и заключается её задача).

Также проверим важность отдельных признаков через вызов `.feature_importances_` различных моделей (для большей равномерности значений, некоторые из них умножим на 100, а некоторые - поделим на 30):

```python
feature_importances = pd.DataFrame(index=X_train.columns.values,
                                   data={
                                       'RandomForestRegressor': rfr_model.feature_importances_*100,
                                       'CatBoostRegressor': cat_model.feature_importances_,
                                       'LGBMRegressor': lgb_model.feature_importances_/30})
```

Рассмотрим их в виде графика:

```python
feature_importances.plot(kind='barh',
                         figsize=(17,10),
                         title='Feature importances',
                         width=0.8)
plt.show()
```

Исходя из данного графика, важность признаков была определена разными моделями по разному:

- Модель `RandomForestRegressor()` определила как наиболее важные признаки - признаки **lag_1** и **hour**, затем уже **rolling_mean**;
- Модель `CatBoostRegressor()` определила как наиболее важные признаки - признаки **hour**, затем уже **rolling_mean** и **lag_1**;
- Модель `LGBMRegressor()` построил своё предсказание преимущественно на отстающих значениях **lag_1**, **lag_12**, **lag_23** и **rolling_mean**;

### Вывод

- Результат достигнут, значение метрики RMSE ниже требуемой (48):
  - Наилучший результат в данном случае показал CatBoostRegressor(), со своим значением RMSE = 41.93;
  - Самый быстрый результат, проходящий проверку на адекватность и удовлетворяющий условиям - LinearRegression() с результатом RMSE = 46.57;
  - **Оптимальной моделью для внедрения** является `LGBMRegressor()`, как самая быстрая (в 11 раз быстрее чем CatBoost), и со схожим результатом RMSE = **42.47**;
- Наибольшую важность признаков у выбранной модели имеют добавленные признаки **отстающих значений** (lag_1 - lag_23);
- График сезонности заказов показывает чёткую взаимосвязь по дням недели, при этом не показывая этого на графике важности признаков, следовательно - это наблюдение ошибочно, и данная взаимосвязь отсутствует.
