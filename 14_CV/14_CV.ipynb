Импорт библиотек необходимых, для выполнения кода:

from tensorflow.keras.layers import GlobalAveragePooling2D, Dense

Фукнции для загрузки тренировочных и тестовых данных, в том числе с выполнением аугментации (смещение и зеркальное отражение по горизонтали)

```python

def load_train(path):
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        horizontal_flip=True,
        width_shift_range=0.2,
        height_shift_range=0.2)
    labels = pd.read_csv(path + 'labels.csv')
    train_datagen_flow = train_datagen.flow_from_dataframe(
        dataframe=labels,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=32,
        class_mode='raw',
        subset='training',
        seed=13) 
    return train_datagen_flow
    

def load_test(path):
    test_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2)
    labels = pd.read_csv(path + 'labels.csv')
    test_datagen_flow = test_datagen.flow_from_dataframe(
        dataframe=labels,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=32,
        class_mode='raw',
        subset='validation',
        seed=13) 
    return test_datagen_flow

```

Функция для создания модели на основе "костяка" ResNet50 с добавлением нескольких полносвязных слоёв.


```python

def create_model(input_shape):
    optimizer = Adam(lr=0.0001)

    backbone = ResNet50(input_shape=input_shape,
                    weights='/datasets/keras_models/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5', 
                    include_top=False)

    #backbone.trainable = False
    model = Sequential()
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(160, activation='relu'))
    model.add(Dense(80, activation='relu'))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(1, activation='relu')) 

    model.compile(optimizer=optimizer, loss='mean_squared_error',
                  metrics=[MeanAbsoluteError()])
    return model


```

Финальная функция по обучению модели с 13ю эпохами и размером батча 16:

```python

def train_model(model, train_data, test_data, batch_size=16, epochs=13,
                steps_per_epoch=None, validation_steps=None):
    if steps_per_epoch is None:
        steps_per_epoch = len(train_data)
    if validation_steps is None:
        validation_steps = len(test_data)
    
    model.fit(train_data,
              validation_data=test_data,
              epochs=epochs,
              steps_per_epoch=steps_per_epoch,
              validation_steps=validation_steps,
              verbose=2, shuffle=True)

    return model

```

````python
baseline_y_true = labels['real_age'].sample(frac=0.2, random_state=13)
baseline_y_pred = pd.Series(baseline_y_true.median(), index=baseline_y_true.index)
````

````python
print('Baseline true values:', baseline_y_true.head(), sep='\n')
print()
print('Baseline predictions:', baseline_y_pred.head(), sep='\n')
print()
print('Baseline MAE:', mean_absolute_error(baseline_y_true, baseline_y_pred))
````

Следовательно, наша модель проходит проверку на адекватность, так как значение $MAE_{model} {\leq} MAE_{baseline}$

- Результатитвность данной модели для использования зависит от её назначения:
  - При необходимости **оценить возрастную группу человека** - ошибка в 5.6 лет вполне **допустима**, к примеру человека 35 лет и 40 - вполне можно отнести к одной группе "взрослые", и погрешность в 5 лет ничего в данном случае не изменит;

- Следовательно, данная модель выполняет только **одну из двух** задач.
