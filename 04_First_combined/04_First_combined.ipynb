### Преобразование данных в нужные типы.

Однозначно не соответсвует тип только у колонки `user_score`, который мы меняем на числовой.

Но для замены необходимо избавиться от существующего строкового значения ***tbd***, с заменой его на NaN.

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan).astype(float)
```

### Обработка пропусков:

Первым делом удалим невосстановимое - строки с пустым значением **name**, они же являются строками с пропущенным значением **genre**. С учётом того, что данные строки составляют менее 0.01% всей выборки, потерь не будет.

```python
data.loc[data['name'].isna()]
```

```python
data.dropna(subset=['name'], inplace=True)
```

Пропуски в годе выпуска игры **year_of_release** обработаем по платформе, на которую они выпускались с использованием среднего значения `mean()`. Обосновано это тем, что игры на какую-либо платформу выпускались преимущественно в срок жизни этой платформы.

```python
data['year_of_release'] = data.groupby('platform')['year_of_release'].transform(lambda x: x.fillna(x.mean())).astype(int)
```

Пропуски в колонке **rating** заполним маркерным значением `absent` ("отсутствует"). Значение RP (Rating Pending), как мне кажется, использовать в данном случае нельзя, хоть и фактически это так же означает отсутствие рейтинга, но это значение уже присутствует в наборе данных.

```python
data['rating'].fillna('absent', inplace=True)
```

```python
pd.DataFrame(round((data.isna().mean()*100),2)).style.background_gradient('coolwarm')
```

Наличие рейтинга, что пользовательского, что критиков необязательно, тем более с отсутствием обязательной централизованной системы рейтингов на протяжении всей временной шкалы исследуемых данных.

### Продажи платформ по годам. Срок их жизни.

Для определения наибольших суммарных продаж выделим воспользуемся `groupby()` и выберем из него "топ 10"

```python
top_10 = data.groupby('platform')[['total_sales']].agg(['sum']).stack()\
.sort_values(
    'total_sales', ascending=False)\
    .head(10)\
    .reset_index()\
    .drop(columns='level_1'
)
```

```python
top_10
```

```python
#data.query('platform == "DS" and year_of_release == 1985')
```

Соберём сводную таблицу с данными по платформам, годам и продажам:

Чтобы защититься от выбросов вроде "игра на Nintnedo DS 1985 года", при выходе консоли в 2004, и на основе выше установленных выводов, ограничимся данными с 1992 года, что так же улучшит читаемость графиков.

```python
sales = data.query('year_of_release >= 1992').pivot_table(
    index=['platform', 'year_of_release'],
    values=['total_sales'],
    aggfunc=['max', 'count']
)

sales.columns = ['total_sales', 'count']
sales.head()
```

И воспользуемся полученными выше данными для построения сводного графика продаж от времени.

```python
sns.set(rc={'figure.figsize':(17,8)})

bx = sns.lineplot(
    data=sales.query('platform in @top_10.platform').reset_index(),
    x="year_of_release",
    y="total_sales",
    hue="platform",)


bx.set_title('График общих продаж по годам')
bx.set_xlabel("Год")
bx.set_ylabel("Число продаж, млн. шт.")
plt.show()
```

**Промежуточный вывод**. По данному графику мы можем сказать, что:

- Средний срок жизни платформы составляет от 8 до 10 лет.
- PC, как платформа не имеет такой характеристики, как срок жизни, а существует параллельно, без сильных всплесков и падений.
- Из актуальных платформ в список "Топ 10" попала только PS4, остальные к 2015 году снизили свои продажи до минимальных значений.

###  Определение потенциально прибыльных платформ.

## Проверка гипотез

</center>

Произведём проверку двух гипотез о равенстве средних двух совокупностей:

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Установим значение критического уровня статистической значимости равное 5%

```python
alpha = 0.05
```

### Гипотеза о рейтингах платформ

Проверим гипотезы с использованием возможностей метода `scipy.stats.ttest_ind`. Значение `equal_var` установим как *True*, так как выборка одна.

Нулевой гипотезой в данном случае станет "*Средние пользовательские рейтинги платформ Xbox One и PC одинаковые*"

Альтернативной же гипотезой будет "*Средние пользовательские рейтинги платформ Xbox One и PC отличаются*"

```python
rating_hypo = st.ttest_ind(
    actual_data.query("platform == 'XOne'")['user_score'].dropna(),
    actual_data.query("platform == 'PC'")['user_score'].dropna(),
    equal_var = True)
 
print('p-значение:', rating_hypo.pvalue)
 
if rating_hypo.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, принимаем альтернативную.')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Следовательно, принимаем нулевую гипотезу как основную, и предполагаем, что средние пользовательские рейтинги платформ Xbox One и PC **одинаковы**

Шаг необязателен, но оценим графически те же два распределения.

```python
xone_ax = actual_data.query("platform == 'XOne'").plot(
    kind='hist',
    y='user_score',
    histtype='stepfilled',
    bins=70,
    linewidth=5,
    color="g",
    alpha=0.7,
    title='Гистограмма оценок пользователей по платформам',
    grid=True,
    figsize=(17,4)
)
pc_ax = actual_data.query("platform == 'PC'").plot(
    kind='hist',
    y='user_score',
    histtype='stepfilled',
    bins=70,
    linewidth=5,
    color="b",
    alpha=0.25,
    ax=xone_ax,
    figsize=(17,4)
)

xone_ax.set_xlabel("Средняя оценка пользователей") 
xone_ax.set_ylabel("Частота значений")
xone_ax.legend(['XOne', 'PC']) 
plt.show()
```

В результате потдвердим результат оценки гипотезы, графически распределение выглядит весьма схоже.

---
### Гипотеза о рейтингах жанров

Проверим ещё две гипотезы тем же методом:

Нулевой гипотезой в данном случае станет "*Средние пользовательские рейтинги жанров Action и Sports одинаковы*"

Альтернативной (в данном случае, интересующей) гипотезой будет "*Средние пользовательские рейтинги жанров Action и Sports разные*"

```python
genre_hypo = st.ttest_ind(
    actual_data.query("genre == 'Action'")['user_score'].dropna(),
    actual_data.query("genre == 'Sports'")['user_score'].dropna(),
    equal_var = True)
 
print('p-значение:', genre_hypo.pvalue)
 
if genre_hypo.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, принимаем альтернативную.')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Следовательно, принимаем альтернативную гипотезу, как основную, и делаем вывод, что средние пользовательские рейтинги жанров Action и Sports **отличаются**.

Аналогично, построим график для проверки

```python
action_ax = actual_data.query("genre == 'Action'").plot(
    kind='hist',
    y='user_score',
    histtype='stepfilled',
    bins=70,
    linewidth=5,
    color="g",
    alpha=0.25,
    title='Гистограмма оценок пользователей по жанрам',
    grid=True,
    figsize=(17,4)
)
sport_ax = actual_data.query("genre == 'Sports'").plot(
    kind='hist',
    y='user_score',
    histtype='stepfilled',
    bins=70,
    linewidth=5,
    color="r",
    alpha=0.75,
    ax=action_ax,
    figsize=(17,4)
)

action_ax.set_xlabel("Средняя оценка пользователей") 
action_ax.set_ylabel("Частота значений")
action_ax.legend(['Action', 'Sports']) 
plt.show()
```

Графически видны разнцица в распределении, что лишний раз отвергает нулевую гипотезу.
